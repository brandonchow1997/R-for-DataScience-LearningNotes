## 1.3 图形属性映射

#### 按类赋予不同颜色

```r
ggplot(data=mpg) + geom_point(mapping=aes(x=displ, y=hwy, color=class))
```

#### 按类赋予不同的散点大小

```r
ggplot(data=mpg) + geom_point(mapping=aes(x=displ, y=hwy, size=class))
```

#### 按类赋予不同的透明度

```r
ggplot(data=mpg) + geom_point(mapping=aes(x=displ, y=hwy, alpha=class))
```

#### 按类赋予不同的形状

```r
ggplot(data=mpg) + geom_point(mapping=aes(x=displ, y=hwy, shape=class))
```

由于ggplot2只能同时使用6种形状，因此多出的变量值不会出现在图中。

## 1.5 分面
将图分割成多个分面

```r
ggplot(data = mpg) + geom_point(mapping = aes(x = displ, y = hwy)) + facet_wrap(~ class, nrow = 2)
# facet_wrap()参数，第一个参数是一个“公式”
```

```r
ggplot(data = mpg) + geom_point(mapping = aes(x = displ, y = hwy)) + facet_grid(drv ~ cyl)
```

## 1.6 几何对象

### 散点图

```r
# geom_point()
ggplot(data = mpg) + geom_point(mapping = aes(x = displ, y = hwy))
```

### 平滑曲线图拟合上图散点

```r
# geom_smooth()
ggplot(data = mpg) + geom_smooth(mapping = aes(x = displ, y = hwy))
```

```r
# geom_smooth()的linetype属性，按照不同的线形绘制不同的曲线
ggplot(data = mpg) + geom_smooth(mapping = aes(x = displ, y = hwy, linetype = drv))
# 此处根据汽车驱动系统（drv）变量的所有值，用3条曲线来表示汽车
```

```r
# 上图中选择不同颜色
# show.legend = FALSE 不添加图例
ggplot(data = mpg) + geom_smooth(mapping = aes(x = displ, y = hwy, linetype = drv, color = drv) ,show.legend = FALSE)
```

### 同一张图中显示多个几何对象

```r
ggplot(data = mpg) + 
geom_point(mapping = aes(x = displ, y = hwy)) + 
geom_smooth(mapping = aes(x = displ, y = hwy))
```

```r
# 效果与上图相同
ggplot(data = mpg, mapping = aes(x = displ, y = hwy)) + 
geom_point() + 
geom_smooth()
```

```r
ggplot(data = mpg, mapping = aes(x = displ, y = hwy)) + 
  geom_point(mapping = aes(color = class)) + 
  geom_smooth()
```

```r
# 筛选出class = "subcompact"的曲线
ggplot(data = mpg, mapping = aes(x = displ, y = hwy)) + 
  geom_point(mapping = aes(color = class)) + 
  geom_smooth(data = filter(mpg, class == "subcompact"), se = FALSE)
```

## 1.7 统计变换
基于ggplot2内置的diamonds数据集

### 条形图

```r
ggplot(data = diamonds) + 
  geom_bar(mapping = aes(x = cut))
```

### 摘要统计量
20多个统计变换（参考ggplot2速查表）

```r
ggplot(data = diamonds) + 
  stat_summary(mapping = aes(x = cut, y = depth),
           fun.ymin = min,
           fun.ymax = max,
           fun.y = median
           )
```

## 1.8 位置调整

```r
# 使用color着色
ggplot(data=diamonds) +
    geom_bar(mapping = aes(x =cut, color = cut))
```

```r
# 使用fill着色
ggplot(data=diamonds) +
    geom_bar(mapping = aes(x =cut, fill = cut))
```

```r
# 将fill图形属性映射到另一个变量（如clarity），条形自动分块堆叠
ggplot(data=diamonds) +
    geom_bar(mapping = aes(x =cut, fill = clarity))
```

```r
# position='fill'，比较组间比例
ggplot(data=diamonds) +
    geom_bar(mapping = aes(x =cut, fill = clarity),
            position='fill')
```
